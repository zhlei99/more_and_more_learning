```python
#5æŠ˜äº¤å‰éªŒè¯
from sklearn.model_selection import KFold
n_splits=5
kfold = KFold(n_splits=n_splits, shuffle=True, random_state=42)  #æ‰“ä¹±ï¼Œéšæœºç§å­
train_x = data[data['target']!=-1][feats]
train_y = data[data['target']!=-1]['target']
res=data[data['target']==-1][['id']]

test_x= data[data['target']==-1][feats]
res['target'] = 0
for train_idx, val_idx in kfold.split(train_x):
    model.random_state = model.random_state + 1   #æ¯æ¬¡éšæœºçŠ¶æ€æ”¹å˜ç§å­
    train_x1 = train_x.loc[train_idx]
    train_y1 = train_y.loc[train_idx]
    test_x1 = train_x.loc[val_idx]
    test_y1 = train_y.loc[val_idx]
    #,(vali_x,vali_y)
    model.fit(train_x1, train_y1,eval_set=[(train_x1, train_y1),(test_x1, test_y1)],eval_metric='auc',early_stopping_rounds=100)
    #æ¯æ¬¡æ¨¡åž‹é¢„æµ‹çš„æ¦‚çŽ‡éƒ½è¿›è¡Œç´¯åŠ ï¼Œæœ€åŽæ±‚å‡å€¼
    res['pred'] += model.predict_proba(test_x)[:,1]

res['target'] = res['target']/n_splits
filepath = './submission' + str(n_splits) + '_fold.csv'
res.to_csv( filepath, index=False)
```
