## 用mlxtend.frequent_patterns进行频繁项集挖掘

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
pd.options.display.max_columns=100
```

```python
#转化为one-hot编码
basket_hot_encoding = pd.get_dummies(basket, prefix='', prefix_sep='')
```

```python
len(basket_hot_encoding.columns)
```

```python
basket_hot_encoding.columns.unique()
```

```python
#找到napkis列值为1 的行。
gg = basket_hot_encoding['napkins']
gg[gg.max(axis=1).isin([1])]
```

```python
gg.groupby(level=0, axis=1).max(axis =1)
```

```python
#列名有1269维，去重复后120维，所以要对数据进行处理
#相同的列名的数值取最大值，并合并成一列
basket_hot_encoding_unique = basket_hot_encoding.groupby(level=0, axis= 1).max(axis = 1)
```

```python
#检查
basket_hot_encoding_unique.head()
```

```python
basket_hot_encoding_unique.columns
```

```python
#将数据标准化
def encode_units(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1
basket_hot_encoding = basket_hot_encoding.applymap(encode_units)
```

```python
#挖掘频繁项集与关联规则
frequent_itemsets = apriori(basket_hot_encoding_unique, min_support=0.02, use_colnames=True)
rules2 = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.05)
```

```python
rules2
```

```python
print("频繁项集：", frequent_itemsets)
print("关联规则：", rules2[ (rules2['lift'] >= 1) & (rules2['confidence'] >= 0.05) ])
```
