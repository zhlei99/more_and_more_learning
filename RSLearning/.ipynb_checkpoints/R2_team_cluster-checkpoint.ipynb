```python
from sklearn.cluster import KMeans
from sklearn import preprocessing
import pandas as pd
import numpy as np
```

```python
#输入数据
data = pd.read_csv('../datasets/team_cluster_data.csv', encoding = 'gbk')
```

```python
#查看前五行
data[0:5]
```

```python
#查看列
data.columns
```

```python
train_x = data[["2019国际排名","2018世界杯排名","2015亚洲杯排名"]]
```

```python
train_x
```

```python
#数据处理归一化,[0,1]
min_max_scaler = preprocessing.MinMaxScaler()
train_x = min_max_scaler.fit_transform(train_x)
```

```python
train_x
```

```python
#创建模型
kmeas = KMeans(n_clusters=3)
kmeas.fit(train_x)
```

```python
predict_y = kmeas.predict(train_x)
```

```python
##合并聚类结果，插入到原始数据中
result = pd.concat([data,pd.DataFrame(predict_y)], axis= 1)
result.rename({0:u'聚类结果'}, axis = 1, inplace = True)  #inplace原表操作
```
