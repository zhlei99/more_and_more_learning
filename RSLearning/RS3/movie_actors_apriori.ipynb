```python
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
#数据加载
movies = pd.read_csv('../../datasets/MovieActors/movie_actors.csv')
```

```python
movies.head()
```

```python
# 将genres进行one-hot编码（离散特征有多少取值，就用多少维来表示这个特征）
movies_hot_encoding = movies.drop(columns='actors', axis= 1).join(movies.actors.str.get_dummies('/'))
```

```python
movies_hot_encoding.head(1)
```

```python
pd.Series.str.get_dummies?
```

```python
#将movieId, title设置为index
movies_hot_encoding.set_index(['index'], inplace=True)
```

```python
# 挖掘频繁项集，最小支持度为0.02
itemsets = apriori(movies_hot_encoding,use_colnames=True, min_support=0.05)
# 按照支持度从大到小进行时候粗排
itemsets = itemsets.sort_values(by="support" , ascending=False) 
```

```python
itemsets
```

```python
#根据频繁项集计算关联规则，设置最小提升度为2
rules = association_rules(itemsets, metric='lift', min_threshold=2)
rules = rules.sort_values(by = 'lift', ascending = False)
rules
```
