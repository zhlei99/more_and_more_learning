```python
#删除SenderPersonId为空的行，可以看到MetadateTo还是有空行，
email_file_reindex.drop(index=nan_index).info()
```

```python
#删除任何有空行的值
email_file_reindex.dropna(inplace=True)
email_file_reindex.info()
```

```python
#处理MetadataFrom列，通过Personid进行统一，
#将email与person文件根据Id合并
email_file_reindex_join = email_file_reindex.join(persons_file.set_index('Id'), on='SenderPersonId')

email_file_reindex_join.rename(columns={'Name':'from_name'}, inplace=True)
email_file_reindex_join
```

```python
email_file_reindex_join['MetadataTo'].unique()
```

根据观察需要做以下清洗工作:  
1、';H' 替换    
2、所有名字小写,并删除掉逗号  
3、所有提取的名字若出现在别名文件里，则对应着人物ID，去关联人物表  

```python
#1、';H' 替换
email_file_reindex_join.loc[3, 'MetadataTo'] = 'H'
email_file_reindex_join.loc[3, 'MetadataTo']
```

```python
#2、所有名字小写,并删除掉逗号
email_file_reindex_join['MetadataTo'] = email_file_reindex_join.MetadataTo.str.lower().str.replace(',', '',regex=False)
```

```python
#3、所有提取的名字若出现在别名文件里，则通过别名关联aliases表，同时通过用户ID，关联
#用户的名字
email_file_reindex_join['to_name'] = email_file_reindex_join.join(aliases_file.set_index('Alias'), on='MetadataTo').join(persons_file.set_index('Id'), on='PersonId')['Name']
```

```python
email_file_reindex_join['to_name']
```

## 至此email_file_reindex_join就是处理好的文件。以下通过此文件处理所需要的格式三元组（u, v , weight)

```python
email_file_reindex_join.columns
```

```python
#增加一个权重列，权重是等于发邮件的次数
grouped = email_file_reindex_join['SenderPersonId'].groupby(by = [email_file_reindex_join['from_name'], email_file_reindex_join['to_name']]).agg(['count'])    
```

```python
#得到一个三元组的数据
grouped.reset_index(inplace= True)
```

```python
grouped.head()
#迭代处理,转换成格式
edges_weights = list(grouped.apply(lambda x: (x['from_name'], x['to_name'], x['count']), axis=1))
```

```python
edges_weights
```

# 搭建算法

```python
#创建有向图
graph = nx.DiGraph()
#批量加入有向图结点权重
graph.add_weighted_edges_from(edges_weights)
# 计算简化模型的PR值
pagerank = nx.pagerank(graph)
```

```python
pagerank
```

```python
# 画网络图
def show_graph(graph, type = 'spring_layout'):
    if type == 'spring_layout':
        # 使用Spring Layout布局，类似中心放射状
        positions=nx.spring_layout(graph)
    if type == 'circular_layout':
        # 使用Circular Layout布局，在一个圆环上均匀分布
        positions=nx.circular_layout(graph)

    # 设置网络图中的节点大小，大小与pagerank值相关，因为pagerank值很小所以需要*20000
    nodesize = [x['pagerank']*20000 for v,x in graph.nodes(data=True)]
    # 设置网络图中的边长度
    edgesize = [np.sqrt(e[2]['weight']) for e in graph.edges(data=True)]
    # 绘制节点
    nx.draw_networkx_nodes(graph, positions, node_size=nodesize, alpha=0.4)
    # 绘制边
    nx.draw_networkx_edges(graph, positions, edge_size=edgesize, alpha=0.2)
    # 绘制节点的label
    nx.draw_networkx_labels(graph, positions, font_size=10)
    # 输出希拉里邮件中的所有人物关系图
    plt.show()
```

```python
# 将pagerank数值作为节点的属性
nx.set_node_attributes(graph, name = 'pagerank', values=pagerank)
# 画网络图图1
show_graph(graph)

# 将完整的图谱进行精简
# 设置PR值的阈值，筛选大于阈值的重要核心节点
pagerank_threshold = 0.005
# 复制一份计算好的网络图
small_graph = graph.copy()
# 剪掉PR值小于pagerank_threshold的节点
for n, p_rank in graph.nodes(data=True):
    if p_rank['pagerank'] < pagerank_threshold: 
        small_graph.remove_node(n)
# 画网络图图2
show_graph(small_graph, 'circular_layout')
```
