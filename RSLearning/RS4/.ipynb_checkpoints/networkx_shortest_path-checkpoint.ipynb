networkx 图论库，含有图论的方法

```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.read_gml('../../datasets/shortest_path/football.gml')
```

```python
G
```

```python
nx.draw(G, with_labes = True)
```

```python
nx.shortest_path(G, source='Buffalo', target='Rice')
```

```python
#Dijkstra算法
print(nx.single_source_dijkstra_path(G, 'Buffalo'))
```

```python
print(nx.multi_source_dijkstra_path(G, {'Buffalo','Rice'}))
```

```python
# Flody算法
print(nx.floyd_warshall(G, weight='weight'))
```
